# Introduction <a name="introduction"></a>

This project is inspired by the ambiguity of basketball positions, based on how the game is played in National-Basketball-Association (NBA) regulation basketball, and the aptness of this question as a classification exercise. In NBA Basketball, there are no players consistently on offense or defense, unlike Football/Soccer and American Football and Baseball. Instead, when possession of the ball changes teams, the same players who were just on defense immediately begin playing offense, and vice-versa. This makes the role of each player harder to understand for someone watching basketball for the first time. In fact, many first-time basketball viewers may not realize that there are positions at all, and junior/recreational basketball teams frequently play without positions. The goal of our work here is to suggest whether these positions exist largely as formalities, or if <u>players' positions can be classified by their past gameplay statistics</u>.

## Loading and Preparing the Data

### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Loading The Season Statistics

To begin our analysis, we need data! We break this process into 4 steps:
1. Load the raw data from our [webscraping library](#appendix1)
2. Filter the raw data to remove:
    * rows where `MP == 0` (minutes played). We will later divide each predictor column by `MP` so it can't be zero.
    * rows where `Year` is outside the 2000-2015 range.
    * rows where `Pos` is not one of the 5 common positions in Basketball today (see [Introduction](#intro))
3. Select the columns we are interested in (see [Methods](#methods))
4. Rename non-predictor columns to be lowercase, and convert the `Pos` column from type character-string to factor. 

```r
# raw_season_data <- get_season_data(2000:2020)
# head(raw_season_data)

# PREVIOUS:
#
urlfile_1 <- "https://raw.githubusercontent.com/cass12345/ddssccii/main/Seasons_Stats.csv"
raw_season_data <- read_csv(url(urlfile_1), show_col_types = FALSE)
```

<br>

There are many columns, so we've printed all the columns below. We will only be selecting the columns that we described in the [Methods](#methods) section.

```r
cat("ALL COLUMNS IN RAW DATAFRAME:\n", colnames(raw_season_data), sep="\t")

# PREVIOUS:
```

<br>

Below, we filter out any rows with zero minutes played, and with a non-standard position (that is, not one of the 5 positions discussed in the [Introduction](#intro)).

```r
season_data <- raw_season_data |>
    filter(MP != 0) |>
    filter(Pos %in% c("PF","PG","SF","SG","C")) |>
    filter(Year >= 2000 & Year <= 2015)
```

<br>

Then, we select the columns that we stated we would work with [above](#methods) (using `select`).

```r
season_data <- season_data |>
    select(Year, Player, Pos, BLK, AST, TRB, STL, MP, "3PA", PF, Tm)
```

<br>

Now, we rename the `Player` and `3PA` columns to be clearer, and change our class column `Pos` to be a `factor` type instead of a `string`:

```r
season_data <- season_data |>
    rename(name = Player) |>
    rename(three_PA = "3PA") |>
    mutate(pos = as_factor(Pos))
```



<br>



### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Wrangling The Season Statistics

At this point, we have selected the columns that we are interested in and filtered for values that will obstruct our analysis, but we face a more complicated question: <u>how do we deal with players who play for multiple teams within one year, or who play different positions and for different teams, throughout their career?</u> We illustrate this problem below by considering two players: Will Barton and LeBron James. First we will consider the case of Will Barton:

```r
will_barton <- season_data |>
    filter(name == "Will Barton")
will_barton
```

<br> 

We can see from the table above that Will Barton played the same position (SG) for three different teams in 2015. If we are to accurately use Barton's statistics to determine his position, we have to add up his statistics across each team. This might be impossible if we were working with derived numbers (ratios, percentages, etc.) but all of our predictor columns are raw counted observations. Below we group the table by name and position to determine Barton's statistics:

```r
will_barton <- will_barton |>
    group_by(name) |>
    summarize(across(.cols  = c(BLK, AST, TRB, STL, MP, three_PA, PF), 
                     .fns   = list(sum=sum),
                     .names = "{.col}"),
              .groups = "keep")
will_barton
```

<br>

So, for a player like Will Barton who plays for multiple teams within one year, the method above (grouping by player's name) to condense statistics works well. What about a player like LeBron James, who has played for multiple positions across his career?

```r
leBron_james <- season_data |>
    filter(name == "LeBron James")
leBron_james
```

<br>

We can see above that LeBron played 3 positions from 2004 to 2015. Should we add all his statistics together from across years? No! We must delineate James' statistics when he is playing different positions. We will do this by including `Pos` in our `group_by` call.

```r
leBron_james <- leBron_james |> 
    group_by(name, pos) |>
    summarize(across(.cols  = c(BLK, AST, TRB, STL, MP, three_PA, PF), 
                     .fns   = list(sum=sum),
                     .names = "{.col}"),
              .groups = "keep")
leBron_james
```

<br>

The table above shows 3 rows for LeBron James, even though he is just a single human being. Is this okay? We've decided it is, because each row corresponds to James in a different position, and we've hypothesized that gameplay statistics are related to position, rather than player. We discuss this idea further in the [Discussion](#discussion). 

However, there is another problem: James' statistics are higher across all columns when he plays SF (Small-Forward), because he has spent the most years playing in this position. We can see this easily by looking at the `MP`  (minutes played) column and see that he has played SF for almost 6 times as long as he played PF (power forward) and more than 10 times as long as he has played SG (shooting guard.) And what about when we consider other players? Will Barton, for example, only played 5 seasons from 2000 to 2015, while LeBron James played 11. To account for these differences in play time by row, either between players or between a single player's different positions, we will divide every predictor column by `MP` (minutes - played). 
<br>

```r
rbind(leBron_james, will_barton) |>
    mutate(across(.cols  = c(BLK, AST, TRB, STL, three_PA, PF),
              .fns   = list(function(.){return(./MP)}),
              .names = "{.col}_per_MP"),
       .keep = "unused")
```

<br> 

We see above that grouping by both name and position lets us separate an individual player's statistics by the different positions they've played throughout their career. This is how we will treat our whole dataset.

```r
season_data <- season_data |>
    group_by(name, pos) |>
    summarize(across(.cols  = c(BLK, AST, TRB, STL, MP, three_PA, PF), 
                     .fns   = list(sum=sum),
                     .names = "{.col}"),
              .groups = "keep") |>
    mutate(across(.cols  = c(BLK, AST, TRB, STL, three_PA, PF),
                  .fns   = list(function(.){return(./MP)}),
                  .names = "{.col}_per_MP"),
           .keep = "unused")
```

### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Loading Player Height, Weight, and Merging Dataframes

So far we've loaded and cleaned the season statistics for each player in the NBA from 2000 to 2015. We're almost done with finalizing our dataset, we just need to match the position height and weight to each player.

```r
# NEW:
# raw_player_data <- get_all_players()


# PREVIOUS:
#
urlfile_2 <- "https://raw.githubusercontent.com/cass12345/ddssccii/main/player_data.csv"
raw_player_data <- read_csv(url(urlfile_2), show_col_types = FALSE)
head(raw_player_data)
```

<br>

Wrangling this data will be easier. All we have to do is convert the height from the "<feet>-<inches>" format to a numerical format, and select the columns we are interested in. Whether we use inches or centimeters doesn't matter since we will be normalizing each height and weight observation in our classification recipe. We convert the height string to a number by separating the feet and inches into their own columns, then redefining `height` using `mutate` to equal `(feet * 12) + inches` .

```r
player_data <- raw_player_data |>
    separate(col  = height,
             into = c("feet","inches"),
             sep  = "-",
             convert = TRUE)|>
    mutate(height = feet * 12  + inches) |>
    select(name, weight, height)

head(player_data)
```

<br>

Now we just have to merge the dataframes.

```r
final_dataset <- merge(x = season_data, y = player_data, by.x = "name", by.y = "name") |>
    drop_na()

head(final_dataset)
```

<br><br><br>

### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Splitting the Train and Test Data

Splitting our data into separate testing and training dataframes will be important for us to test the accuracy of our classification model after we have made it.

```r
data_split <- initial_split(final_dataset, prop = 0.75, strata = pos)
player_train <- training(data_split)
player_test <- testing(data_split)
```

```r
ball_recipe1 <- recipe(pos ~ BLK_per_MP + AST_per_MP + TRB_per_MP + three_PA_per_MP + weight + height, 
                       data = player_train) |>
    step_center(all_predictors()) |>
    step_scale(all_predictors())


ks <- tibble(neighbors = seq(from = 80, to = 100, by = 1))

ball_vfold1 <- vfold_cv(player_train, v = 5, strata = pos)

ball_spec1 <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
    set_engine("kknn") |>
    set_mode("classification")

ball_fit1 <- workflow() |>
    add_recipe(ball_recipe1) |>
    add_model(ball_spec1) |>
    tune_grid(resamples = ball_vfold1, grid = ks) |>
    collect_metrics()

ball_fit1
```

## Settling: predicting Guards and Forwards

```r
guard_or_forward <- function (position) {
    print(position)
    if (toString(position) %in% c("C", "SF", "PF")) {
        return("F")
    } else {
        return("G")
    }
}

player_train |>
    rowwise() |>
    mutate(x = guard_or_forward(pos)) |>
    mutate(pos = as_factor(pos))
```

```r
head(player_train)
```

```r
forwards <- player_train |>
    filter(pos %in% c("C", "SF", "PF")) |>
    mutate(pos = "G")

guards <- player_train |>
    filter(pos %in% c("PG", "SG")) |>
    mutate(pos = "F")

f_g_data <- rbind(forwards, guards)
```

```r
ball_recipe1 <- recipe(pos ~ BLK_per_MP + AST_per_MP + TRB_per_MP + three_PA_per_MP + weight + height, 
                       data = f_g_data) |>
    step_center(all_predictors()) |>
    step_scale(all_predictors())


ks <- tibble(neighbors = seq(from = 2, to = 150, by = 1))

ball_vfold1 <- vfold_cv(f_g_data, v = 5, strata = pos)

ball_spec1 <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
    set_engine("kknn") |>
    set_mode("classification")

ball_fit1 <- workflow() |>
    add_recipe(ball_recipe1) |>
    add_model(ball_spec1) |>
    tune_grid(resamples = ball_vfold1, grid = ks) |>
    collect_metrics()

```
